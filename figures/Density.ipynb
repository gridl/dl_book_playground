```python
# load required libraries:
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('default')
plt.figure(figsize=(6,6))
plt.style.use('default') # macht den hintergrund weg 
plt.rcParams.update({'font.size': 15}) #Macht grosse axen
import seaborn as sns
import pandas as pd
```

```python
np.random.seed(4)
params = {'mathtext.default': 'regular' }          
plt.rcParams.update(params)

mean, cov = [0, -1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 75)

mean, cov = [3, 1], [(1, -.5), (-.5, 0.5)]
data2 = np.random.multivariate_normal(mean, cov, 75)

data = np.vstack((data,data2))

df = pd.DataFrame(data, columns=["x", "y"])
sns.kdeplot(df.x, df.y)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.scatter(df.x,df.y, c='r', edgecolors='black')
plt.text(-4,3,'$p_θ(x_1,x_2)$')
plt.savefig('ch06_den.pdf')
```
